```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_excel("C:/Users/htcso/Downloads/EVDS (4).xlsx")
```

```python
data["Date"] = pd.to_datetime(data["Date"], format = "%d-%m-%Y")
```

```python
data.info()
```

```python
data.set_index("Date", inplace = True)
```

```python
data.head()
```

```python
data_weekly = data.resample(rule = "W").mean().round(4) # convert frequency to weekly
data_weekly
```

```python
data_monthly = data.resample(rule = "M").mean().round(4) # convert frequency to monthly
data_monthly
```

```python
data_quarter = data.resample(rule = "Q").mean().round(4) # convert frequency to quarter
data_quarter
```

```python
data_yearly = data.resample(rule = "Y").mean().round(4) # convert frequency to yearly
data_yearly
```
